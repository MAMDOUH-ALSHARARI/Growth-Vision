## 2.2


```python
total_saudis = last_year_quarter["Number_Of_Saudis"].sum()
```

```python
male_female_per_eco = (
    last_year_quarter[
        ["Economic_Activity", "Number_Of_Male_Saudis", "Number_Of_Female_Saudis"]
    ]
    .groupby(by="Economic_Activity")
    .sum()
    .reset_index()
)
male_female_per_eco["total_saudis"] = total_saudis
male_female_per_eco
```

```python
# calculate percentage for male and female
male_female_per_eco["Male_Percentage"] = (
    (male_female_per_eco["Number_Of_Male_Saudis"] / total_saudis) * 100
).round(2)
male_female_per_eco["Female_Percentage"] = (
    (male_female_per_eco["Number_Of_Female_Saudis"] / total_saudis) * 100
).round(2)
male_female_per_eco["Rate_Difference"] = abs(
    (male_female_per_eco["Male_Percentage"] - male_female_per_eco["Female_Percentage"])
).round(2)
```

```python
male_female_per_eco.head()
```

```python
male_female_per_eco[
    male_female_per_eco["Economic_Activity"]
    == "أنشطة المكاتب الرئيسية، والأنشطة الاستشارية في مجال الإدارة"
]["Rate_Difference"].values.tolist()[0]
```

```python
male_female_per_eco.to_csv("data/male_female_per_eco.csv", index=False)
```

```python
male_female_per_region_eco = (
    last_year_quarter[
        [
            "region",
            "Number_Of_Male_Saudis",
            "Number_Of_Female_Saudis",
            "Economic_Activity",
        ]
    ]
    .groupby(by=["region", "Economic_Activity"])
    .sum()
    .reset_index()
)
male_female_per_region_eco["total"] = (
    male_female_per_region_eco["Number_Of_Male_Saudis"]
    + male_female_per_region_eco["Number_Of_Female_Saudis"]
)
male_female_per_region_eco.head()
```

#### filter cities that is larger than 1000 total saudi employees


```python
male_female_per_region_eco = male_female_per_region_eco[
    male_female_per_region_eco["total"] > 2000
]
```

#### calculate rate diff between male/female each economy activity


```python
male_female_per_region_eco["Male_Percentage"] = (
    (
        male_female_per_region_eco["Number_Of_Male_Saudis"]
        / male_female_per_region_eco["total"]
    )
    * 100
).round(2)
male_female_per_region_eco["Female_Percentage"] = (
    (
        male_female_per_region_eco["Number_Of_Female_Saudis"]
        / male_female_per_region_eco["total"]
    )
    * 100
).round(2)
male_female_per_region_eco["Rate_Difference"] = abs(
    (
        male_female_per_region_eco["Male_Percentage"]
        - male_female_per_region_eco["Female_Percentage"]
    )
).round(2)
```

```python
male_female_per_region_eco.head()
```

```python
male_female_per_region_eco = male_female_per_region_eco[
    male_female_per_region_eco["Rate_Difference"] != 0
]
```
